```
import numpy as np
import matplotlib.pyplot as plt

n_samples = 100

# 平均(-3, 1)、標準偏差(2, 4) の散布図を描く
X = np.random.normal((-3, 1), (2, 4), (n_samples, 2))

plt.xlim(-10, 10)
plt.ylim(-10, 10)
plt.scatter(X[:, 0], X[:, 1])
```

```
mu = X.mean(axis=0) # 平均μ
var = X.var(axis=0) # 分散σ^2 (標準偏差σの二乗)
eps = 1e-6 # 小さい正の定数ε

gamma = 1 # パラメータγ
beta = 0 # パラメータβ

norm_X = (X - mu) / np.sqrt(var + eps)
norm_X = gamma * norm_X + beta

plt.xlim(-10, 10)
plt.ylim(-10, 10)
plt.scatter(X[:, 0], X[:, 1])
plt.scatter(norm_X[:, 0], norm_X[:, 1])

print(f'μ={norm_X.mean(axis=0)}, σ={norm_X.var(axis=0)}')
```
