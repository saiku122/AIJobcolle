```
import numpy as np
import matplotlib.pyplot as plt

def tanh(x):
    return np.tanh(x)

def sigmoid(x):
    return 1 / (1 + np.exp(-x))

def relu(x):
    return np.maximum(0, x)

def swish(x):
    return x * sigmoid(x)

def hswish(x):
    return x * (np.minimum(relu(x+3), 6) / 6)

def softplus(x):
    return np.log(1 + np.exp(x))

def mish(x):
    return x * np.tanh(softplus(x))

def elu(x):
    alpha = 1
    return np.where(x > 0.0, x, alpha * (np.exp(x) - 1))

X = np.arange(-10, 10, 0.01)

plt.figure(figsize=(15, 5))
plt.xlim(-8, 8)
plt.ylim(-1.0, 1.1)
plt.plot(X, sigmoid(X), label="Sigmoid")
plt.plot(X, tanh(X), label="Tanh")
plt.plot(X, softplus(X), label="Softplus")
plt.plot(X, relu(X), label="ReLU")
plt.plot(X, swish(X), label="Swish")
plt.plot(X, hswish(X), label="Hard Swish")
plt.plot(X, mish(X), label="Mish")
plt.plot(X, elu(X), label="ELU")
plt.legend()
```
